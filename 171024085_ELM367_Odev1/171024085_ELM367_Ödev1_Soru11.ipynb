```python
import math
import matplotlib.pyplot as plt
import numpy as np

# Jupyter Notebook'ta şekillerin görünmesini sağlamak için gerekli:
%matplotlib notebook 

alpha = math.pi/2
```

```python
n_a = np.arange(-20,30,1)
x_a = np.zeros(len(n_a)) # a şıkkı için hafıza x'e n adedince yer ayrıldı
wo_a = 0.1*math.pi # a şıkkı için wo değeri

for k in range(len(n_a)):
    x_a[k] = math.sin(wo_a*n_a[k] + alpha)
    #A seçeneğinin çizdirilmesi
plt.figure()
plt.stem(n_a,x_a, use_line_collection=True)
plt.title("$sin(\omega_0n+\\alpha),\\alpha = \pi/2, \omega_0 = 0.1\pi$")
plt.show()
```
