```python
import math
import matplotlib.pyplot as plt
import numpy as np
# Jupyter Notebook'ta şekillerin görünmesini sağlamak için gerekli:
%matplotlib notebook


#2.7 a için çizim
pi = np.pi
n_a = np.arange(0,30)
angle_a = n_a*np.pi/6
x_a = np.exp(1j*angle_a)
    
plt.stem(n_a, x_a, use_line_collection=True)
plt.title('$Soru7)--2.7.a) e^{j(\pi*n/6)}$')
plt.show()
```

```python
n_d = np.arange(0,30)
x_d = np.exp(1j*n_d)

plt.figure()
plt.stem(n_d, x_d, use_line_collection=True)
plt.title('$Soru7)--2.40.d) e^{jn}$')
plt.show()
```
