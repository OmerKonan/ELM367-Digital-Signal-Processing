```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib notebook 
n_ = np.arange(-10,10,1)
h_a = np.zeros(len(n_)) 

# Soru 6) 18-a
for n in range(len(n_)):
    h_a[n] = (0.5 ** n_[n]) if (n_[n]>=0) else 0
    
plt.stem(n_, h_a,  use_line_collection=True)
plt.title("$ h[n] = (1/2)^n u[n]$")
plt.show()
```

```python
# Soru 6) 18-c

h_c = np.zeros(len(n_)) 

for n in range(len(n_)):
    h_c[n] = (0.5 ** abs(n_[n]))
plt.figure()
plt.stem(n_, h_c,  use_line_collection=True)
plt.title("$ h[n] = (1/2)^{|n|}$")
plt.show()
```

```python
# Soru 6) 18-e

h_e = np.zeros(len(n_))
h1 = np.zeros(len(n_)) 
h2 = np.zeros(len(n_)) 

for n in range(len(n_)):
    
    h1[n] = ((1/3)**n_[n]) if(n_[n] >= 0) else 0
    h2[n] = (3.**n_[n]) if(-(n_[n])-1>=0) else 0
    h_e[n] = h1[n] + h2[n]
    
plt.figure()
plt.stem(n_, h_e,  use_line_collection=True)
plt.title("$ h[n] = (1/3)^n u[n] + 3^n u[-n-1]$")
plt.show()
```
