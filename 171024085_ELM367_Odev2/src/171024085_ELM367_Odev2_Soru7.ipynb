```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib notebook 


n_a = np.arange(-4,4,1)
h_a = np.zeros(len(n_a)) 

# Soru 7) 19-a
for n in range(len(n_a)):
    h_a[n] = (4 ** n_a[n]) if (n_a[n]>=0) else 0
    #print("n:",n_[n])
    #print("h",h_a[n])
    
plt.stem(n_a, h_a,  use_line_collection=True)
plt.title("$ h[n] = 4^n u[n]$")
plt.show()
```

```python
pi = np.pi
n_e = np.arange(-20,20,1)
h_e = np.zeros(len(n_e)) 

# Soru 7) 19-e
for n in range(len(n_e)):
    h_e[n] = ((3/4) ** abs(n_e[n]))*np.cos(n_e[n]*pi/4+pi/4) 
    #print("n:",n_[n])
    #print("h",h_a[n])

plt.figure()    
plt.stem(n_e, h_e,  use_line_collection=True)
plt.title("$ h[n] = (3/4)^{|n|} cos(\\pi n/4 + \\pi/4)$")
plt.show()
```
