```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib notebook 
# Soru 8) 21-c
n_c = np.arange(-3,8,1)
h_c = np.array([0,0,1,1,1,1,1,0.5,0,0,0])

n_c = n_c/2

for k  in range(len(n_c)):
    
    if (n_c[k]%1 != 0):
        n_c[k] = 0

plt.stem(n_c, h_c, use_line_collection=True)
plt.title("h[n] = 2n")
plt.show()
```
