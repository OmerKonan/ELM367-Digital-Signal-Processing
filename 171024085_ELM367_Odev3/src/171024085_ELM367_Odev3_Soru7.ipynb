```python
import math
import matplotlib.pyplot as plt
import numpy as np
# Jupyter Notebook'ta şekillerin görünmesini sağlamak için gerekli:
%matplotlib notebook
```

```python

w = np.linspace(-np.pi, np.pi, 10000)
H = 1 - (0.45* np.exp(-1j*w))/(1-0.8*np.exp(-1j*w))


#H = (1 - (1.25 * np.exp(-1j * w))) / (1 - (0.8 * np.exp(-1j * w)))

plt.plot(w, np.abs(H), color='b', label='Genlik')
plt.plot(w, np.angle(H), color='g', label='Faz')

plt.legend()

plt.show()
```
